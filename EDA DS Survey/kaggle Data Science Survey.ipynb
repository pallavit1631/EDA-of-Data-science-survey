# Primary tool used at work/school to analyze data?

```python
df2=data[['Q12_MULTIPLE_CHOICE']].groupby(['Q12_MULTIPLE_CHOICE']).size().reset_index(name="Number_of_respondent")
df2.columns=['Multiple_choice','Number_of_respondent']
```

```python
df2=df2.drop(axis=0,index=6)
df2
```

```python
df2=df2.sort_values(by='Number_of_respondent',ascending=True)
```

```python
fig, ax = plt.subplots(figsize=(15,7), dpi= 90)
ax.vlines(x=df2['Multiple_choice'], ymin=0,ymax=df2['Number_of_respondent'], color='red', alpha=0.7, linewidth=2)
plt.xticks(rotation=270)

ax.scatter(x=df2['Multiple_choice'], y=df2['Number_of_respondent'], s=20, color='black', alpha=0.7)

# Title, Label, Ticks and Ylim
ax.set_title('Primary tool used at work/school to analyze data:', fontdict={'size':10})
ax.set_ylabel('Number_of_respondent')
ax.set_xticklabels(df2['Multiple_choice'].str.upper(), fontdict={'horizontalalignment': 'right', 'size':8})
ax.set_xticks(df2.Multiple_choice)
# Annotate
for row in df2.itertuples():
    ax.text(row.Multiple_choice, row.Number_of_respondent+150, s=(row.Number_of_respondent), horizontalalignment= 'center', verticalalignment='bottom', fontsize=12)

plt.show()
```
